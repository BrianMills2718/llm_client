# Runtime Durability Follow-Ups From grounded-research

**Status:** planned  
**Repo:** `llm_client`  
**Trigger:** grounded-research needed app-local mitigations to finish long benchmark runs.


## Contract

**Current:**
- app-local `LLM_CLIENT_DB_PATH` isolation to avoid SQLite lock failures
- app-local finite request timeouts for long structured calls
- manual DB inspection to diagnose a trace

**Target:**
1. concurrent observability writes are durable enough for normal multi-run use
2. long structured-call timeout policy is shared runtime behavior
3. one trace can be summarized across `llm_calls` and `tool_calls`


## Pre-Made Decisions

- build on existing SQLite + JSONL logging
- keep per-project overrides, but make them optional mitigation
- durability first, diagnostics second, default cleanup third
- do not reopen cancelled default-centralization work as pure refactor


## Acceptance Questions

After this slice, a human should be able to answer:
- Did concurrent logging fail?
- What was the last successful LLM call in this trace?
- What tool calls ran under the same trace?
- Did a long structured call time out, retry, or complete?
