# Gemini Strict-Schema Behavior Study

Goal: collect run-backed evidence for how Gemini behaves on Tyler-like structured-output schemas.


## Input Contract

- Model list: one or more Gemini models
- Fixed case registry: flat, nested, optional, list-of-objects, enum-heavy schemas
- Prompt per case: deterministic instruction to fill the schema faithfully


## Output Contract

- Per-case status
- execution_path (`native_schema`, `instructor`, or `error`)
- response_format_type
- validation_errors
- aggregate JSON summary suitable for downstream prompt_eval analysis
